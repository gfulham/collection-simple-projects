# Fractals in Python
## The Barnsley Fern
 ![image](400_Barnsleys_Fern.png)
 
This project aims to use matplotlib to recreat the Barnsley Fern. The [fern](https://en.wikipedia.org/wiki/Barnsley_fern)
is one of the basic examples of self-similar sets, i.e. it is a mathematically generated pattern that can be reproducible at any magnification or reduction. Like the Sierpinski triangle, the Barnsley fern shows how graphically beautiful structures can be built from repetitive uses of mathematical formulas with computers.

This is a guided project from a python youtuber named [Python Programmer](https://www.youtube.com/channel/UC68KSmHePPePCjW4v57VPQg). The short video may be found [here](https://www.youtube.com/watch?v=pfMNFD-y4rM&ab_channel=PythonProgrammer)

```python
# libaries
import numpy as np # for choosing the functions a random
import matplotlib.pyplot as plt # for plotting the fern
import matplotlib.cm as cm # just adding color palettes
%matplotlib inline
```

```python
def func_1(x,y):
    return (0, 0.16*y)
def func_2(x,y):
    return (.85*x + 0.04*y, -0.04*x + 0.85*y + 1.6)
def func_3(x,y):
    return (0.2*x - 0.26*y, 0.23*x + 0.22*y + 1.6)
def func_4(x,y):
    return (-0.15*x + 0.28*y, 0.26*x + 0.24*y + 0.44)

functions = [func_1, func_2, func_3, func_4]
```

```python
# set variables
points = 100_000
x, y = 0, 0
x_list = []
y_list = []

# create function 
for i in range(points):
    function = np.random.choice(functions, p = [0.01, 0.85, 0.07, 0.07])
    x, y = function(x,y)
    x_list.append(x)
    y_list.append(y)
```

```python
# simple plot using matplotlib
plt.scatter(x_list, y_list, s=0.25, color = 'green')
```
